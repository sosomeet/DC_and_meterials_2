```python
import pandas as pd
import numpy as np
from pathlib import Path
from scipy.stats import skew, kurtosis

# ============================================================
# Descriptive Statistics (논문 Table 1 대응)
# ============================================================

# -----------------------------
# Config
# -----------------------------
BASE_DIR = Path("./")

INPUT_FILE = BASE_DIR / "tvpvar_input_scaled.csv"
OUTPUT_FILE = BASE_DIR / "descriptive_stats.csv"

DATE_CANDIDATES = ["Date", "date", "DATE"]

# -----------------------------
# Load
# -----------------------------
df = pd.read_csv(INPUT_FILE)

# 날짜 컬럼 제거
date_col = None
for c in DATE_CANDIDATES:
    if c in df.columns:
        date_col = c
        break

if date_col:
    df = df.drop(columns=[date_col])

# numeric 컬럼만 사용
df = df.select_dtypes(include=[np.number])

print("Using columns:", list(df.columns))

# -----------------------------
# Descriptive Statistics
# -----------------------------
rows = []

for col in df.columns:
    s = df[col].dropna()

    row = {
        "variable": col,
        "mean": s.mean(),
        "std": s.std(),
        "min": s.min(),
        "max": s.max(),
        "skewness": skew(s),
        "kurtosis": kurtosis(s, fisher=True),  # excess kurtosis
        "n_obs": len(s)
    }

    rows.append(row)

result = pd.DataFrame(rows)

# -----------------------------
# 정렬 (논문 스타일)
# -----------------------------
result = result[
    ["variable", "mean", "std", "min", "max", "skewness", "kurtosis", "n_obs"]
]

# -----------------------------
# 저장
# -----------------------------
result.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")

print("\nSaved:", OUTPUT_FILE)
print("\nPreview:")
print(result.to_string(index=False))
```
