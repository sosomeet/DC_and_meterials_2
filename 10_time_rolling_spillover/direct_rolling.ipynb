```python
import pandas as pd
import matplotlib.pyplot as plt

# ============================
# rolling spillover 결과 불러오기
# ============================

df = pd.read_csv("./rolling_spillover.csv")

if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])

# ============================
# NET 계산
# ============================

df["SOLVPN_NET"] = df["SOLVPN_TO"] - df["SOLVPN_FROM"]
df["COPPER_NET"] = df["COPPER_TO"] - df["COPPER_FROM"]

# ============================
# 1) Rolling SOLVPN TO / FROM
# ============================

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["SOLVPN_TO"], label="SOLVPN_TO")
plt.plot(df["Date"], df["SOLVPN_FROM"], label="SOLVPN_FROM")
plt.title("Rolling SOLVPN Directional Spillover")
plt.xlabel("Time")
plt.ylabel("Spillover (%)")
plt.legend()
plt.tight_layout()
plt.savefig("rolling_solvpn_directional.png", dpi=300)
plt.show()

# ============================
# 2) Rolling COPPER TO / FROM
# ============================

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["COPPER_TO"], label="COPPER_TO")
plt.plot(df["Date"], df["COPPER_FROM"], label="COPPER_FROM")
plt.title("Rolling COPPER Directional Spillover")
plt.xlabel("Time")
plt.ylabel("Spillover (%)")
plt.legend()
plt.tight_layout()
plt.savefig("rolling_copper_directional.png", dpi=300)
plt.show()

# ============================
# 3) Rolling Net Spillover
# ============================

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["SOLVPN_NET"], label="SOLVPN_NET")
plt.plot(df["Date"], df["COPPER_NET"], label="COPPER_NET")
plt.axhline(0, linewidth=1)
plt.title("Rolling Net Spillover")
plt.xlabel("Time")
plt.ylabel("Net Spillover (%)")
plt.legend()
plt.tight_layout()
plt.savefig("rolling_net_spillover.png", dpi=300)
plt.show()
```
