```python
import pandas as pd
from pathlib import Path

# ============================================================
# Config
# ============================================================

BASE_DIR = Path("./")

FILES = [
    "da_hrl_lmps_2022.csv",
    "da_hrl_lmps_2023.csv",
    "da_hrl_lmps_2024_part1.csv",
    "da_hrl_lmps_2024_part2.csv",
    "da_hrl_lmps_2025.csv",
    "da_hrl_lmps_2026.csv",
]

OUTPUT_DIR = Path("../")
OUTPUT_DIR.mkdir(exist_ok=True)

OUTPUT_FILE = OUTPUT_DIR / "da_hrl_lmps_2022_2026_concat.csv"

# ============================================================
# Load + Concat
# ============================================================

dfs = []

for file_name in FILES:
    file_path = BASE_DIR / file_name

    print(f"[LOAD] {file_path}")

    df = pd.read_csv(file_path)

    print(f" -> shape: {df.shape}")

    dfs.append(df)

# concat
merged_df = pd.concat(dfs, axis=0, ignore_index=True)

# ============================================================
# Datetime 정렬
# ============================================================

datetime_candidates = [
    "datetime_beginning_ept",
    "datetime_beginning_utc",
    "Datetime",
    "DATE",
    "Date",
]

date_col = None

for col in datetime_candidates:
    if col in merged_df.columns:
        date_col = col
        break

if date_col is not None:
    print(f"[INFO] Using datetime column: {date_col}")

    merged_df[date_col] = pd.to_datetime(
        merged_df[date_col],
        errors="coerce"
    )

    merged_df = merged_df.sort_values(date_col).reset_index(drop=True)

# ============================================================
# Save
# ============================================================

merged_df.to_csv(OUTPUT_FILE, index=False)

print("\n================================================")
print("[DONE]")
print(f"Saved file: {OUTPUT_FILE}")
print(f"Final shape: {merged_df.shape}")
print("================================================")
```
